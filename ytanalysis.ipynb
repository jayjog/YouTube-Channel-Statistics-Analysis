```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


data_size = 100
df = pd.DataFrame({
    'category': np.random.choice(['Tech', 'Cooking', 'Travel', 'Gaming'], size=data_size),
    # Engagement rate between 0.01 and 0.15 (1% to 15%)
    'engagement_rate': np.random.uniform(0.01, 0.15, size=data_size),
    'length_category': np.random.choice(['Short', 'Medium', 'Long'], size=data_size),
    # Views with a wide range, ideal for log scale
    'views': np.random.randint(500, 5000000, size=data_size),
})

# Display the first few rows to confirm data structure
print("--- Sample DataFrame Head ---")
print(df.head())
print("\n")



plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='category', y='engagement_rate', estimator='mean')
plt.title('Average Engagement Rate by Category')
plt.xticks(rotation=45)
plt.show()


plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='length_category', y='views')
plt.title('Views Distribution by Video Length')
plt.yscale('log')  # Log scale for wide range
plt.show()


plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='views', y='engagement_rate', hue='category', alpha=0.7)
plt.title('Views vs. Engagement Rate by Category')
plt.xscale('log')
plt.show()
```
